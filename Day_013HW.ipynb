* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」
* ③ 試著爬爬看其他版的文章

```python
import requests
from bs4 import BeautifulSoup

r = requests.get('https://www.ptt.cc/bbs/NBA/index.html')
soup = BeautifulSoup(r.text,'html5lib')
print(soup)
```

```python
#sample code
time_stamp = []
for item in soup.find_all(class_ = 'title'):
    print(item.text.replace('\t', '').replace('\n', ''))
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+item.find('a')['href']).text, "html5lib")
        print('作者: ' + r.find(class_='article-meta-value').text)
        for tag in r.find_all(class_ ='article-metaline'):
            if tag.find(class_ = 'article-meta-tag').text == '時間':
                time_stamp.append(tag.find(class_ = 'article-meta-value').text)
                print('發表時間:{}'.format(tag.find(class_ = 'article-meta-value').text))
    except:
        continue
```

```python
from datetime import datetime
import time
import calendar
stamps = []
pa = '%b %d %H:%M:%S %Y'
for i in time_stamp:
    stamps.append(datetime.strptime(,pa))
print(datetime.strptime('Nov 29 00:46:01 2019',pa))
```
