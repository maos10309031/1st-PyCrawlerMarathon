```python
import requests 
from bs4 import BeautifulSoup
# 爬取第一頁
r = requests.get('https://movies.yahoo.com.tw/movie_intheaters.html')
soup = BeautifulSoup(r.text , 'html5lib')
releases = soup.find_all(class_ = 'release_info')
for re in releases:
    print('*電影名:{} '.format(re.find(class_ = 'gabtn').text.replace('\n','').replace(' ','')))
    print(' 期待度:{} ,滿意度:{}'.format(re.find('dt').find('span').text,re.find('dd').find('span')['data-num']))
    print('劇情介紹:\n{}'.format(re.find(class_ = 'release_text').find('span').text))
```

```python
options = soup.find(class_ = 'top_change').find_all('option')
movie_id = {}
for op in options:
    movie_id[op.text] = op['value']
```

```python
#爬取一到五頁
for i in range(5):
    r = requests.get('https://movies.yahoo.com.tw/movie_intheaters.html?page={}'.format(i+1))
    soup = BeautifulSoup(r.text,'html5lib')
    releases = soup.find_all(class_ = 'release_info')
    for re in releases:
        print('*電影名:{} '.format(re.find(class_ = 'gabtn').text.replace('\n','').replace(' ','')))
        print(' 期待度:{} ,滿意度:{}'.format(re.find('dt').find('span').text,re.find('dd').find('span')['data-num']))
        print('劇情介紹:\n{}'.format(re.find(class_ = 'release_text').find('span').text))
```
