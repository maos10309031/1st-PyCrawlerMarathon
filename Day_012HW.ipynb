```python
import requests
from bs4 import BeautifulSoup
import re
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
# print(r.text)
regex = '2020/01/02.*'
pattern = re.compile(regex)
soup = BeautifulSoup(r.text,'html5lib')
iterable = soup.find(class_ = 'part_list_2')
items = iterable.find_all('h3')
#滾動模組以下的內容不知如何使用~
for item in items:
    print(item)
```

```python
for item in items:
    print(item.find(class_ = 'date').text,item.find('a').text)
```

```python
title_dict = dict()
regex = '.*\u3000.*'
pattern = re.compile(regex)
pattern_sub = re.compile('\u3000')
for item in items:
#     print(item)
#     print(type(item.find('em')['class'][1]))
    if not(item.find('em')['class'][1] in title_dict.keys()):
        title_dict[item.find('em')['class'][1]] = []
    title = item.find_all('a')[-1].text
    if re.match(pattern ,title):
        title = re.sub(pattern_sub,' ',title)
    text = '時間:{}標題:{}'.format(item.find(class_ = 'date').text, title)
    title_dict[item.find('em')['class'][1]].append(text)
print(title_dict['c_news'][:10])
```

```python
for category in title_dict:
    print('{:13}:{}'.format(category,len(title_dict[category])))
#c_news 政治板的最多
```
